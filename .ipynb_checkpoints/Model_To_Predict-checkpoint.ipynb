```python
import cv2
import keras
```

```python
import numpy as np
CATEGORIES = ['Cat', 'Dog']


def image(folder):
    img = cv2.imread(folder, cv2.IMREAD_GRAYSCALE)
    new_arr = cv2.resize(img, (60, 60))
    new_arr = np.array(new_arr)
    new_arr = new_arr.reshape(-1, 60, 60, 1)
    return new_arr

model = keras.models.load_model('3x3x64-catvsdog.model')
```

```python
prediction = model.predict([image('sample_copied\cats\cat.2.jpg')])
print(CATEGORIES[prediction.argmax()])
```
